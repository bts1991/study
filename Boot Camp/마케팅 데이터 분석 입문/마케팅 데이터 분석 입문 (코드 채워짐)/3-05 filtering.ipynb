# Filtering
- pd.DataFrame 에서 Filtering 은 조건에 부합(Boolean 값이 True)하는 row 만 남기는 과정
```python
df[조건]
# boolean-series:[True, False, ..., True]
```
  


````python
# 예시
import pandas as pd

# 예제 데이터프레임 생성
data = {
    'Name': ['철수', '영희', '흥민', '범근'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago']
}

df = pd.DataFrame(data)
````

````python
df
````

````python
# df.Age >  30; age_over_30

df
con = df.Age > 30

df['age_over_30'] = con # 새로운 컬럼 추가
````

````python
con
````

````python
df
````

````python
# filter result

df[con] # 조건에 부합하는 로우만 출력
````

````python
df[ df.Age > 30 ]
````

## between(), isin() , isna()
```python

```

````python
#titannic btw age

import seaborn as sns
df = sns.load_dataset('titanic')
````

````python
con = df.age.between(20, 22)
````

````python
df[con]
````

````python
df[con].age.value_counts()
````

````python
#isin embarked

df.embarked.value_counts()
````

````python
con = df.embarked.isin(['S', 'C']) # 리스트로 감싸야
df[con].embarked.value_counts()
````

````python
#isna deck ~ 사용: 반대로

df.deck.value_counts() # 결측치라는 null/NaN 값은 나오지 않음
df.deck.value_counts(dropna = False) # 결측치 포함
````

````python
con = df.deck.isna()
df[con]
````

````python
df[~con] #부정
````

## and(&), or(|)
- 두 개 이상의 조건을 결합
  - and: 모두 True -> True
  - or: 하나라도 True -> True


````python
 # True or False

 True and False
````

````python
True or False
````

````python
True & False # 기호로 대체 가능
````

````python
True | False
````

````python
df = pd.DataFrame({'num' : [1,2,3,4,5,6,7,8,9]})
````

````python
df
````

````python
df['col_1'] = df.num >= 6
df['col_2'] = df.num >= 3
df['col_and'] = df.col_1 & df.col_2
df['col_or'] = df.col_1 | df.col_2
df
````

1. 연습문제: 타이타닉 데이터셋 필터링
- 문제:아래는 타이타닉 호 승객 데이터에 대한 간략한 설명과 함께 필터링 작업을 수행하는 문제입니다. Seaborn 라이브러리를 사용하여 타이타닉 데이터셋을 불러온 후, 다음 작업을 수행하세요.
  - class가 'First'이고, age가 30 이상인 승객들의 데이터를 필터링하세요.
  - survived가 1(생존)이거나 embark_town이 'Southampton'인 승객들의 데이터를 필터링하세요.
  - fare가 100 이상이고, sex가 'female'인 승객들의 데이터를 필터링하세요.



````python
import seaborn as sns
df = sns.load_dataset('titanic')
````

````python
df.head()
````

````python
df.info()
````

````python
# class가 'First'이고, age가 30 이상인 승객들의 데이터를 필터링하세요.
````

````python
df[df['class'] == 'First'][df['age'] > 30][['class', 'age']]
````

````python
# survived가 1(생존)이거나 embark_town이 'Southampton'인 승객들의 데이터를 필터링하세요.
````

````python
df['survived'].value_counts()
df['embark_town']
````

````python
con1 = df['survived'] == 1
con2 = df['embark_town'] == 'Southampton'
df[con1 | con2][['survived','embark_town']]
````

````python
# fare가 100 이상이고, sex가 'female'인 승객들의 데이터를 필터링하세요.
````

````python
con3 = df['fare'] >= 100
con4 = df['sex'] == 'female'
df[con3 & con4][['fare','sex']]
````

````python
# class가 'First'이고, age가 30 이상인 승객들의 데이터를 필터링하세요.

filter1 = (df['class'] == 'First') & (df.age >= 30 )
filter1
````

````python
df[filter1].age.value_counts().sort_index()
````

````python
df[filter1]['class'].value_counts()
````

````python
# survived가 1(생존)이거나 embark_town이 'Southampton'인 승객들의 데이터를 필터링하세요.

filter2 = (df.survived == 1) | (df.embark_town == 'Southampton')
filter2
````

````python
df2 = df[filter2]
````

````python
df2[df2.survived != 1].value_counts()
````

````python
df2[df2.survived != 1].embark_town.value_counts()
````

````python
# fare가 100 이상이고, sex가 'female'인 승객들의 데이터를 필터링하세요.

fil3 = (df.fare >= 100) & (df.sex == 'female')
df3 = df[fil3]
````

````python
df3.fare.value_counts().sort_index()
````

````python
df3.sex.value_counts()
````
